```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
def scrapweb(url):
    driver = webdriver.Chrome()
    driver.get(url)
#     driver.implicitly_wait(5)
    soup = BeautifulSoup(driver.page_source,'lxml')
    tables = soup.find_all('table')
    dfs = pd.read_html(str(tables))[0]
    input_tag = driver.find_element_by_id('big_table_next')
    nextButtonClass =(input_tag.get_attribute('class'))
    while("paginate_button_disabled" not in nextButtonClass):
        driver.find_element_by_id('big_table_next').click()
        soup = BeautifulSoup(driver.page_source,'lxml')
        tables = soup.find_all('table')
        df = pd.read_html(str(tables))
        dfs =dfs.append(df[0])
        nextButtonClass=driver.find_element_by_id('big_table_next').get_attribute('class')
    driver.close()
    return dfs
```

```python
dfs = scrapweb('https://www.bankeauctions.com/')
print(dfs)
```
